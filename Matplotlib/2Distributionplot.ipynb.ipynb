```python
import seaborn as sns
import matplotlib.pyplot as plt
```


We use distribution plots to visualize the distribution of **quantitative** data (can't be categorized).

Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:

- distplot/histplot
- jointplot
- pairplot
- rugplot

```python
df = sns.load_dataset('tips')
df
```

```python
df['sex'].unique() # qualitive data (can be Categories)
```

```python
df['tip'].unique() #quantitative data
```

```python
df['total_bill'].unique() #quantitative data
```

**distplot/histplot**

```python
sns.histplot(df['total_bill'],bins = 20,kde = True)
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins = 20,kde = True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins = 20,kde = True)
```

**jointplot**

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df,kind = 'scatter')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df,kind = 'reg')
```

**pairplot**

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex',palette='rainbow')
sns.pairplot(df,hue='size',palette='rainbow')
```

**rugplot**

```python
sns.rugplot(df)
```

```python
sns.rugplot(df['tip'])
```
