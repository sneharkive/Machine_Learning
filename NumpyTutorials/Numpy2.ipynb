# **Array Manipulation**

Stacking Array (You cannot stack two Matrix, you need vector(array) for it)

There are three stack
1. Vertical Stack
2. Horizontal Stack
3. Simple Stack

```python
a = np.array([1, 2, 3, 4])
b = np.array([5, 6, 7, 8])
```

```python
np.vstack((a, b)) #Vertical
```

```python
np.hstack((a, b)) #Horizontal
```

```python
np.column_stack((a, b)) #Column Stack
```
