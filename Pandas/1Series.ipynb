# **Pandas Series**
A series is a 1D labeled array capable of holding any data type. The aixs labels are collectively called the index.

```python
import numpy as np
import pandas as pd
```

```python
labels = ['a', 'b', 'c']
my_list = [10, 20, 30]
arr = np.array([10, 20, 30])
d = {'p':10, 'q':20, 'r':30}
```

```python
pd.Series(my_list)
```

```python
# Set Custom Index
pd.Series(my_list, index=labels)
```

```python
pd.Series(arr)
```

```python
pd.Series(d)
```
