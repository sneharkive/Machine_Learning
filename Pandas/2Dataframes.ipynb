# **Creating a DataFrame**

```python
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}
df = pd.DataFrame(data)
```

```python
df
```

```python

data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]
df2 = pd.DataFrame(data_list)
```

```python
df2
```

```python
columns = ["Name","Age","City","Salary"]
df2 = pd.DataFrame(data_list,columns =columns)
df2
```

Selection and Indexing of Columns

```python
df2[["Name","City"]]
```

Creating a new column

```python
df2["Designation"] = ["Doctor","Eng.","Doctor","Eng."]
```

```python
df2
```

Removing Columns

```python
df2.drop(1,axis = 0) # Not gonna change anything in real df2
```

```python
df2  #no change
```

```python
df2.drop(1,axis = 0, inplace=True) # It's gonna make changes in real df2
```

```python
df2
```

```python
# Multiple row/column can be delete
df2.drop(["Designation", "City"],axis = 1, inplace=True)
```
